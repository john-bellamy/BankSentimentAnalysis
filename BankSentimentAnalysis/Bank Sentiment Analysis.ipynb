```python
import vaderSentiment.vaderSentiment as vs
import pandas as pd
from pandas import DataFrame
```

```python
tweets =  pd.read_csv('Data\\bank_tweets.csv')
```

```python
tweets.head()
```

```python
analyzer = vs.SentimentIntensityAnalyzer()
```

```python
def sent_fun(word):
    """ Uses VaderSentiment and assigns a default of blank if an error occurs."""
    try:
        s = analyzer.polarity_scores(word)
    except: 
        s = ''
    return s
```

```python
boa = tweets[tweets['Bank Name'] == 'Bank of America']
```

```python
boa.tail()
```

```python
wells = tweets[tweets['Bank Name'] == 'Wells Fargo Bank']
```

```python
chase = tweets[tweets['Bank Name'] == 'Chase Bank']
```

```python
boa_sent_list = []
for x in range(len(boa['Clean-Tweets'])):
    boa_sent_list.append(sent_fun(boa['Clean-Tweets'][x]))
```

```python
boa_sent_list[0:10]
```

```python
boa_sent_frame = DataFrame.from_records(boa_sent_list, columns=['compound','neg','neu','pos'])
```

```python
chase.tail()
```

```python
boa_sent_frame["Bank"] = 'Bank of America'
```

```python
boa_sent_frame.head()
```

```python
wells_sent_list = []
for x in range(3000,4500):
    wells_sent_list.append(sent_fun(wells['Clean-Tweets'][x]))
```

```python
wells_sent_frame = DataFrame.from_records(wells_sent_list, columns=['compound','neg','neu','pos'])
```

```python
wells_sent_frame["Bank"] = 'Wells Fargo Bank'
```

```python
chase_sent_list = []
for x in range(1500,3000):
    chase_sent_list.append(sent_fun(chase['Clean-Tweets'][x]))
```

```python
len(boa_sent_list)
```

```python
chase_sent_frame = DataFrame.from_records(chase_sent_list, columns=['compound','neg','neu','pos'])
```

```python
chase_sent_frame["Bank"] = 'Chase Bank'
```

```python
sentiment_frame = DataFrame()
```

```python
sentiment_frame = sentiment_frame.append(boa_sent_frame)
```

```python
sentiment_frame = sentiment_frame.append(wells_sent_frame)
```

```python
sentiment_frame = sentiment_frame.append(chase_sent_frame)
```

```python
sentiment_frame.head()
```

```python
sentiment_frame.count()
```

```python
sentiment_frame.head()
```

```python
def get_sentiment(sentiment):
    """Assigns a string sentiment representation of sentiment based on nummeric criteria."""
    if sentiment >= .5:
        return 'pos'
    elif sentiment <= -.5:
        return 'neg'
    else:
        return 'neu'


    
    
    
```

```python
sentiment_frame['Sentiment'] = [get_sentiment(x) for x in sentiment_frame["compound"]]
```

```python
sentiment_frame.head()
```

```python
sent_final =  sentiment_frame[sentiment_frame['Sentiment'] != 'neu']
```
